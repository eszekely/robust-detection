```python
import matplotlib.pyplot as plt
import netCDF4
import numpy as np
import os
import pandas as pd
import random
import scipy
import seaborn as sns
import sys

from mpl_toolkits.basemap import Basemap
from sklearn import linear_model
from sklearn.gaussian_process.kernels import RBF
from sklearn.kernel_ridge import KernelRidge
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import GridSearchCV, KFold, train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from scipy.stats.stats import pearsonr   

sys.path.append("./../src/")
from functions import funcNbModels, funcResolution, repeatModelRadForcing, plotBasemap

%matplotlib inline
```

# Analysis for 1870 - 2005

```python
# Parameters
startDate, endDate, rcp, grid, forcing = 1870, 2100, 85, 20, "anthro"

# Read RCP temperature data
filename_data = "./../output/data/temperature_RCP" + str(rcp) + "_scaling" + str(grid).zfill(2) + "_" + \
                str(startDate) + "_" + str(endDate) + "_" + str(funcNbModels(rcp)) + "models.pkl"
data_df = pd.read_pickle(filename_data)

# Read radiative forcing data
rf = pd.DataFrame()
rf['Observed'] = repeatModelRadForcing(pd.read_pickle("./../output/data/radForcing_RCP" + \
                    str(rcp) + "_" + forcing + "_" + str(startDate) + "_" + str(endDate) + ".pkl"), funcNbModels(rcp))

# Split the data into training and testing
X_train, X_test, y_train, y_test = train_test_split(data_df, rf, shuffle=False, test_size = (endDate-startDate+1)*5)

#param_grid = {'alpha':[1e-4],
#             'kernel':[RBF(gamma) for gamma in np.logspace(-3, 0, 10)]}
regr_grid = [linear_model.LinearRegression(fit_intercept = True), 
             linear_model.LassoCV(max_iter=5000, cv = 5), 
             linear_model.RidgeCV(alphas = train_data.shape[0] * np.logspace(-2,0,100) / 2, cv = 5), 
             linear_model.ElasticNetCV(max_iter=5000, cv = 5),
            KernelRidge(kernel='rbf', gamma = 0.001, alpha = 0.1)]

title = ["Linear regression", "Lasso CV", "Ridge CV", "Elastic net CV", "Kernel regression"]

r2_scores = []

fig, axes = plt.subplots(1, 5, figsize=(18,4))
fig.suptitle("Time: " + str(startDate) + "-" + str(endDate))

for i in range(len(regr_grid)):
    regr = regr_grid[i]
    regr.fit(X_train, y_train.values.ravel())
    y_predicted = regr.predict(X_test)
    r2_scores.append(r2_score(y_test.values.ravel(), y_predicted))

    sns.regplot(x = y_test.values.ravel(), y = y_predicted, 
                marker = ".", color = "blue", line_kws={'color':'black'}, ax = axes[i])
    axes[i].set_title(title[i])
    axes[0].set_ylabel("Predicted")

fig = plt.figure()
r2_scores_df = pd.DataFrame(r2_scores, columns = [forcing])
r2_scores_df.plot(marker = 's')
plt.xticks([0, 1, 2, 3, 4], title, rotation = 45)
plt.ylabel("R2 score")
plt.legend()
```

```python
# Parameters
startDate, endDate, rcp, grid = 1870, 2100, 45, 20

# Read data
filename_data = "./../output/data/temperature_RCP" + str(rcp) + "_scaling" + str(grid).zfill(2) + "_" + \
                str(startDate) + "_" + str(endDate) + "_" + str(funcNbModels(rcp)) + "models.pkl"
data_df = pd.read_pickle(filename_data)

rf = pd.DataFrame()
rf['Observed TOTAL_ANTHRO_RF'] = repeatModelRadForcing(pd.read_pickle("./../output/data/radForcing_RCP" + str(rcp) + \
                    "_" + "anthro" + "_" + str(startDate) + "_" + str(endDate) + ".pkl"), funcNbModels(rcp))
rf['Observed VOLCANIC_RF'] = repeatModelRadForcing(pd.read_pickle("./../output/data/radForcing_RCP" + str(rcp) + \
                    "_" + "volcanic" + "_" + str(startDate) + "_" + str(endDate) + ".pkl"), funcNbModels(rcp))
rf['Observed SOLAR_RF'] = repeatModelRadForcing(pd.read_pickle("./../output/data/radForcing_RCP" + str(rcp) + \
                    "_" + "solar" + "_" + str(startDate) + "_" + str(endDate) + ".pkl"), funcNbModels(rcp))

X_train, X_test, y_train, y_test = train_test_split(data_df, rf, shuffle=False, test_size = (endDate-startDate+1)*3)

# Create a scaler object
sc = StandardScaler(with_mean=False, with_std=False)
# Fit the scaler to the training data and transform
X_train_std = sc.fit_transform(X_train)
# Apply the scaler to the test data
X_test_std = sc.transform(X_test)

train_data = X_train #_std
test_data = X_test #_std

#param_grid = {'alpha':[1e-4],
#             'kernel':[RBF(gamma) for gamma in np.logspace(-3, 0, 10)]}
regr_grid = [linear_model.LinearRegression(), # KernelRidge(), 
             linear_model.MultiTaskLassoCV(max_iter=5000, cv = 5), 
             linear_model.RidgeCV(alphas = train_data.shape[0] * np.logspace(-2,0,10) / 2, cv = 5), 
             linear_model.MultiTaskElasticNetCV(max_iter=5000, cv = 5),
            KernelRidge(kernel='rbf', gamma = 0.001, alpha = 0.1)]
title = ["Linear regression", "Lasso CV", "Ridge CV", "Elastic net CV", "Kernel regression"]

r2_scores = []
r2_scores_uniform_average = []
r2_scores_var_weighted = []

fig, axes = plt.subplots(3, 5, figsize=(18,12), sharey = False)
fig.suptitle("RCP " + str(rcp) + " - Time: " + str(startDate) + "-" + str(endDate))

for i in range(len(regr_grid)):
    regr = regr_grid[i]
    regr.fit(X_train, y_train)
    y_predicted = regr.predict(X_test)
    r2_scores.append(r2_score(y_test, y_predicted, multioutput = 'raw_values'))
    r2_scores_uniform_average.append(r2_score(y_test, y_predicted, multioutput = 'uniform_average'))
    r2_scores_var_weighted.append(r2_score(y_test, y_predicted, multioutput = 'variance_weighted'))
#    filename = "./../output/figures/RCP45_RCP85_" + str(i) + ".jpg"
#    plotBasemap(regr.coef_.reshape(resolution_global20), title[i])

#     sns.regplot(x=y_test, y = y_predicted, marker = ".", color = "blue", 
#                 line_kws={'color':'black'}, ax = axes[i])
#     axes[i].set_title(title[i])

    sns.regplot(x=y_test.iloc[:,0], y = y_predicted[:,0], marker = ".", color = "blue", 
                line_kws={'color':'black'}, ax = axes[0, i])
    axes[0, i].set_title(title[i])
    axes[0, 0].set_ylabel("Predicted TOTAL_ANTHRO_RF")

    sns.regplot(x=y_test.iloc[:,1], y = y_predicted[:,1], marker = ".", color = "blue", 
                line_kws={'color':'black'}, ax = axes[1, i])
    axes[1, 0].set_ylabel("Predicted VOLCANIC_RF")
    sns.regplot(x=y_test.iloc[:,2], y = y_predicted[:,2], marker = ".", color = "blue", 
                line_kws={'color':'black'}, ax = axes[2, i])
    axes[2, 0].set_ylabel("Predicted SOLAR_RF")
    
#fig.savefig("./../output/figures/RCP85_global20_scatterplots_1870_2100.jpg")
#fig.tight_layout()
fig = plt.figure()
r2_scores_df = pd.DataFrame(r2_scores, columns = ['TOTAL_ANTHRO_RF', 'VOLCANIC_RF', 'SOLAR_RF'])
#r2_scores_df = pd.DataFrame(r2_scores, columns = ['VOLCANIC_RF'])
r2_scores_df['Variance weighted'] = pd.DataFrame(r2_scores_var_weighted)
r2_scores_df['Uniform average'] = pd.DataFrame(r2_scores_uniform_average)
r2_scores_df.plot(marker = 's')
plt.xticks([0, 1, 2, 3, 4], title, rotation = 45)
plt.ylabel("R2 score")
plt.legend()
#fig.savefig("./../output/figures/RCP85_global20_r2score.jpg")
print(r2_scores_df)
```
