### Regression with constraints 

- Orthogonality with the residuals: $w^T(y-X\beta)=0$.

- Orthogonality with the signal itself: $w^TX\beta=0$. 

```python
# Create a scaler object
sc_X = StandardScaler(with_mean=True, with_std=True)
sc_y = StandardScaler(with_mean=True, with_std=True)
sc_X_test = StandardScaler(with_mean=True, with_std=True)
sc_y_test = StandardScaler(with_mean=True, with_std=True)

# Fit the scaler to the data and transform
X_train_std = sc_X.fit_transform(X_train.values)
y_train_std = sc_y.fit_transform(y_train.values)
X_test_std = sc_X_test.fit_transform(X_test.values)
y_test_std = sc_y_test.fit_transform(y_test.values)

X = X_train_std
y = y_train.values
Xt = X_test_std
y_test_true = y_test.values

std_X_train = X_train.std(axis=0) # can also get it back from the scaler object
```

```python
constraintForcing = "volcanic"

if constraintForcing == "anthro":
    w = y_ant_std
if constraintForcing == "solar":
    w = y_sol_std
elif constraintForcing == "volcanic":
    w = y_vol_std
elif constraintForcing == "natural":
    w = y_nat_std
```

```python
n, p = X_train.shape

type_constraint = "residuals"
lambdaSeq = np.logspace(0, 5, 5)

rmse_train = np.zeros(len(lambdaSeq))
r2_train = np.zeros(len(lambdaSeq))
rmse_test = np.zeros(len(lambdaSeq))
r2_test = np.zeros(len(lambdaSeq))

mu_vals = np.zeros(len(lambdaSeq))

for i in range(len(lambdaSeq)):
    lambdaVal = X_train_std.shape[0] * lambdaSeq[i] / 2

    mu = np.zeros((1, 1))
    A = np.zeros((p + mu.shape[0], p + mu.shape[0]))
    P = np.zeros((p + mu.shape[0], 1))
    C = np.zeros((p + mu.shape[0], 1))

    # ---- A ----
    A11 = np.matmul(np.transpose(X_train_std), X_train_std) + lambdaVal * np.eye(p)
    A12 = np.matmul(np.transpose(X_train_std), w)
    A21 = np.transpose(A12)
    A22 = np.zeros(mu.shape)

    A[:p, :p] = A11
    A[:p, p:p + 1] = A12
    A[p:p + mu.shape[0], :p] = A21
    A[p:p + mu.shape[0], p:p + 1] = A22

    # ---- C ----
    C[:p] = np.matmul(np.transpose(X_train_std), y_train_std)
    if type_constraint == "signal":
        C[p:p + mu.shape[0]] = np.zeros(mu.shape) # orthogonality with the signal itself
    elif type_constraint == "residuals":
        C[p:p + mu.shape[0]] = np.matmul(np.transpose(w), y_train_std) # orthogonality with the residuals

    # ---- P ----
    A_inv = np.linalg.inv(A)
    P = np.matmul( A_inv, C)
    coefStd_constr = P[:p]
    mu = P[p:]
    mu_vals[i] = mu
    
    y_hat = np.matmul(X_test_std, coefStd_constr)
    rmse_test[i] = np.sqrt(np.mean((y_test_std - y_hat)**2)) 
    r2_test[i] = r2_score(y_test_std, y_hat)

    y_hat_train = np.matmul(X_train_std, coefStd_constr)
    rmse_train[i] = np.sqrt(np.mean((y_train_std - y_hat_train)**2)) 
    r2_train[i] = r2_score(y_train_std, y_hat_train)

plt.semilogx(lambdaSeq, mu_vals, '.-', label = "Lagrange multiplier $μ$")
plt.xlabel("λ", fontsize = 14)
plt.legend(fontsize = 14);
```

```python
fig, axes = plt.subplots(2, 2, figsize = (17, 9))

type_constraint = "residuals" # "residuals" or "signal"
lambda_val = 100

#  ----- No constraint -----
regr = linear_model.Ridge(alpha = X_train_std.shape[0] * lambda_val / 2 )
regr.fit(X_train_std, y_train_std)
y_test_pred = regr.predict(X_test_std)

rmse = np.sqrt(np.mean((y_test_std - y_test_pred)**2)) 
r2 = r2_score(y_test_std, y_test_pred)

fig.sca(axes[0, 0])
plt.plot(y_test_std, label = "True forcing")
plt.plot(y_test_pred, label = "Predicted forcing")
plt.title("No constraint --- $λ = $" + str(lambda_val) + " --- RMSE = " + str(np.round(rmse,4)) + \
              " --- R2 = " + str(np.round(r2,4)))
plt.legend()

fig.sca(axes[0, 1])
coefRaw = np.transpose(regr.coef_) / np.array(std_X_train).reshape(X_train.shape[1], 1)
plotBasemap_subplots(coefRaw.reshape(grid), cLim = None, title_subplot = "No constraint - Raw coefficients");

# ----- With constraint -----
n, p = X_train.shape
mu = np.zeros((1, 1))

A = np.zeros((p + mu.shape[0], p + mu.shape[0]))
P = np.zeros((p + mu.shape[0], 1))
C = np.zeros((p + mu.shape[0], 1))

# ---- A ----
A11 = np.matmul(np.transpose(X_train_std), X_train_std) + X_train_std.shape[0] * lambda_val / 2 * np.eye(p)
A12 = np.matmul(np.transpose(X_train_std), w)
A21 = np.transpose(A12)
A22 = np.zeros(mu.shape)

A[:p, :p] = A11
A[:p, p:p + 1] = A12
A[p:p + mu.shape[0], :p] = A21
A[p:p + mu.shape[0], p:p + 1] = A22

# ---- C ----
C[:p] = np.matmul(np.transpose(X_train_std), y_train_std)
if type_constraint == "signal":
    C[p:p + mu.shape[0]] = np.zeros(mu.shape) # orthogonality with the signal itself
elif type_constraint == "residuals":
    C[p:p + mu.shape[0]] = np.matmul(np.transpose(w), y_train_std) # orthogonality with the residuals

# ---- P ----
A_inv = np.linalg.inv(A)
P = np.matmul( A_inv, C)
coefStd_constr = P[:p]
mu = P[p:]

y_hat = np.matmul(X_test_std, coefStd_constr)
rmse_val = np.sqrt(np.mean((y_test_std - y_hat)**2)) 
r2_val = r2_score(y_test_std, y_hat)

fig.sca(axes[1, 0])
plt.plot(y_test_std, label = "True forcing")
plt.plot(y_hat, label = "Predicted forcing")
if type_constraint == "signal":
    title = "Constraint " + constraintForcing + " : $w^T Xβ = 0$ --- $λ = $" + str(lambda_val) + " --- RMSE = " + \
              str(np.round(rmse_val,4)) + " --- R2 = " + str(np.round(r2_val, 4))
elif type_constraint == "residuals":
    title = "Constraint " + constraintForcing + " : $w^T (y-Xβ) = 0$ --- $λ = $" + str(lambda_val) + " --- RMSE = " + \
              str(np.round(rmse_val,4)) + " --- R2 = " + str(np.round(r2_val, 4))
plt.title(title)
plt.legend()

fig.sca(axes[1, 1])
coefRaw_constr = coefStd_constr / np.array(std_X_train).reshape(X_train.shape[1], 1)
plotBasemap_subplots(coefRaw_constr.reshape(grid), cLim = None, title_subplot = "Constraint " + constraintForcing + " - Raw coefficients");
```
