```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
import netCDF4
from sklearn import linear_model
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ExpSineSquared
from sklearn.kernel_ridge import KernelRidge
from sklearn.model_selection import GridSearchCV

%matplotlib inline
```

### Ordinary least squares and Gaussian processes regression - single model run (GHG and natural forcing)

```python
nc_tas_Nat = netCDF4.Dataset('./data/_DATA/CMIP5/1D/global/annual/tas/tas_mon_GFDL-CM3_historicalNat_r1i1p1_g025.nc')
nc_tas_GHG = netCDF4.Dataset('./data/_DATA/CMIP5/1D/global/annual/tas/tas_mon_GFDL-CM3_historicalGHG_r1i1p1_g025.nc')
```

```python
ncdata_tas_Nat = np.array(nc_tas_Nat.variables['tas'])
ncdata_tas_Nat = ncdata_tas_Nat.reshape(ncdata_tas_Nat.shape[0])

ncdata_tas_GHG = np.array(nc_tas_GHG.variables['tas'])
ncdata_tas_GHG = ncdata_tas_GHG.reshape(ncdata_tas_GHG.shape[0])

times = nc_tas_Nat.variables['time']
dates = pd.to_datetime(netCDF4.num2date(times[:], times.units)).year

ncdata_tas_Nat_df = pd.DataFrame(ncdata_tas_Nat, index = dates, columns = ['Temperature'])
ncdata_tas_GHG_df = pd.DataFrame(ncdata_tas_GHG, index = dates, columns = ['Temperature'])
```

```python
nbPredict = 20
```

```python
ols_GHG = linear_model.LinearRegression()
ols_GHG.fit(np.array(dates[:-nbPredict]).reshape(-1,1), ncdata_tas_GHG[:-nbPredict].reshape(-1,1))

ols_Nat = linear_model.LinearRegression()
ols_Nat.fit(np.array(dates[:-nbPredict]).reshape(-1,1), ncdata_tas_Nat[:-nbPredict].reshape(-1,1))
```

```python
ols_GHG.get_params(deep=True)
```

```python
kernel_gp = RBF(2)
gp_GHG = GaussianProcessRegressor(kernel = kernel_gp, alpha = 1e-6).fit(np.array(dates[:-nbPredict]).reshape(-1,1), ncdata_tas_GHG[:-nbPredict].reshape(-1,1))
gp_Nat = GaussianProcessRegressor(kernel = kernel_gp, alpha = 1e-4).fit(np.array(dates[:-nbPredict]).reshape(-1,1), ncdata_tas_Nat[:-nbPredict].reshape(-1,1))
```

```python
xfit_GHG = dates[:-nbPredict] 
yfit_GHG_OLS = ols_GHG.predict(xfit_GHG[ :, np.newaxis])
yfit_GHG_GP, sigma_GHG_GP = gp_GHG.predict(xfit_GHG[ :, np.newaxis], return_std=True)

xfit_GHG_predict = dates[-nbPredict - 1 :] 
yfit_GHG_OLS_predict = ols_GHG.predict(xfit_GHG_predict[ :, np.newaxis])
yfit_GHG_GP_predict, sigma_GHG_GP_predict = gp_GHG.predict(xfit_GHG_predict[ :, np.newaxis], return_std=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))
ax1.plot(ncdata_tas_GHG_df, label = 'anthropogenic forcing')
ax1.plot(xfit_GHG, yfit_GHG_OLS, 'r', label = 'OLS: slope = ' + '%.4f' % np.asscalar(ols_GHG.coef_))
ax1.plot(xfit_GHG_predict, yfit_GHG_OLS_predict, 'g', label = 'OLS predict')
ax1.plot(xfit_GHG, yfit_GHG_GP, 'k', label = 'Gaussian processes')
ax1.plot(xfit_GHG_predict, yfit_GHG_GP_predict, 'm', label = 'Gaussian processes predict')
ax1.fill(np.concatenate([xfit_GHG_predict, xfit_GHG_predict[::-1]]),
         np.concatenate([yfit_GHG_GP_predict.reshape(-1) - 1.9600 * sigma_GHG_GP_predict,
                        (yfit_GHG_GP_predict.reshape(-1) + 1.9600 * sigma_GHG_GP_predict)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
ax1.set(xlabel = 'Year')
ax1.set(ylabel = 'Temperature (K)')
ax1.set_title('Global mean temperature (GFDL)')
ax1.legend()

xfit_Nat = dates[:-nbPredict] # np.linspace(1870, 2005, 100)
yfit_Nat_OLS = ols_Nat.predict(xfit_Nat[ :, np.newaxis])
yfit_Nat_GP, sigma_Nat_GP = gp_Nat.predict(xfit_Nat[ :, np.newaxis], return_std=True)

xfit_Nat_predict = dates[-nbPredict - 1 :] # np.linspace(1870, 2005, 100)
yfit_Nat_OLS_predict = ols_Nat.predict(xfit_Nat_predict[ :, np.newaxis])
yfit_Nat_GP_predict, sigma_Nat_GP_predict = gp_Nat.predict(xfit_Nat_predict[ :, np.newaxis], return_std=True)

ax2.plot(ncdata_tas_Nat_df, label = 'natural forcing')
ax2.plot(xfit_Nat, yfit_Nat_OLS, 'r', label = 'OLS: slope = ' + '%.4f' % np.asscalar(ols_Nat.coef_))
ax2.plot(xfit_Nat_predict, yfit_Nat_OLS_predict, 'g', label = 'OLS predict')
ax2.plot(xfit_Nat, yfit_Nat_GP, 'k', label = 'Gaussian processes')
ax2.plot(xfit_Nat_predict, yfit_Nat_GP_predict, 'm', label = 'Gaussian processes predict')
ax2.fill(np.concatenate([xfit_Nat_predict, xfit_Nat_predict[::-1]]),
         np.concatenate([yfit_Nat_GP_predict.reshape(-1) - 1.9600 * sigma_Nat_GP_predict,
                        (yfit_Nat_GP_predict.reshape(-1) + 1.9600 * sigma_Nat_GP_predict)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
ax2.set(xlabel = 'Year')
ax2.set(ylabel = 'Temperature (K)')
ax2.set_title('Global mean temperature (GFDL)')
ax2.legend()
#plt.show()
```

### Noisy observations of anthropogenic forcing

```python
# Anthropogenic forcing
dirFiles = "./data/_DATA/CMIP5/1D/global/annual/tas/"
selFileContains = "historicalGHG"
ncdata_df = pd.DataFrame()

for f in os.listdir(dirFiles):
    if selFileContains in f:
        temp_nc = netCDF4.Dataset(dirFiles + f)
        temp_ncdata = np.array(temp_nc.variables['tas']) - np.mean(temp_nc.variables['tas'])
        temp_ncdata = temp_ncdata.reshape(temp_ncdata.shape[0])
        
        temp_ncdata_df = pd.DataFrame(temp_ncdata, columns =[f])
        ncdata_df = pd.concat( [ncdata_df, temp_ncdata_df], axis = 1 )

times = temp_nc.variables['time']
dates = pd.to_datetime(netCDF4.num2date(times[:], times.units)).year
ncdata_df.set_index(dates, inplace = True)
ncdata_df = ncdata_df.drop(ncdata_df.columns[[8,9]], axis = 1) # 8 and 9 have bad outliers

for i in range(ncdata_df.shape[1]): 
    plt.plot(ncdata_df.iloc[:,i])
plt.plot(ncdata_df.mean(axis=1), 'k', lw = 4)
#plt.errorbar([i for i in range(1870,2006)], ncdata_df.mean(axis=1), ncdata_df.var(axis=1), fmt='k.', markersize=10, label='Observations')
plt.ylabel("Temperature anomaly")
plt.title("Anthropogenic forcing")
```

```python
x_fit = np.array(dates).reshape(-1,1)
y_fit_GHG = ncdata_df.mean(axis=1)

start = 1870
end = 2050
x_predict = np.linspace(start, end, end-start).reshape(-1,1)

ols_GHG = linear_model.LinearRegression()
ols_GHG.fit(x_fit, y_fit_GHG)

kernel_gp = RBF(2)
# alpha - value added to the diagonal of the kernel matrix during fitting. Larger values correspond to increased 
# noise level in observations.
gp_GHG = GaussianProcessRegressor(kernel = kernel_gp, alpha = 1e-3).fit(x_fit, y_fit_GHG)

y_predict_GHG_OLS = ols_GHG.predict(x_predict)
y_predict_GHG_GP, sigma_GHG_GP = gp_GHG.predict(x_predict, return_std=True)

fig, (ax1) = plt.subplots(1, 1, figsize=(14, 6))

for i in range(ncdata_df.shape[1]): 
    ax1.plot(ncdata_df.iloc[:,i])
ax1.plot(ncdata_df.mean(axis=1), 'k', lw = 4)

ax1.plot(x_predict, y_predict_GHG_OLS, 'g', lw = 4, label = 'OLS predict')
ax1.plot(x_predict, y_predict_GHG_GP, 'b', lw = 4, label = 'Gaussian processes predict')
ax1.fill(np.concatenate([x_predict, x_predict[::-1]]),
         np.concatenate([y_predict_GHG_GP.reshape(-1) - 1.9600 * sigma_GHG_GP,
                        (y_predict_GHG_GP.reshape(-1) + 1.9600 * sigma_GHG_GP)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
ax1.set(xlabel = 'Year')
ax1.set(ylabel = 'Temperature anomaly (K)')
ax1.set_title('Global mean temperature anomaly')
#ax1.legend()
plt.show()
```

```python
start = 1870
end = 2100
x_predict = np.linspace(start, end, end-start).reshape(-1,1)

# Gaussian processes
param_grid = {"alpha": [1e0, 1e-1, 1e-2, 1e-3], 
              "kernel": [RBF(l) for l in np.logspace(0,3,10)]}
gp = GridSearchCV(GaussianProcessRegressor(), cv = 5, param_grid = param_grid)
gp_GHG = gp.fit(x_fit, y_fit_GHG)
# kernel_gp = RBF(2)
# alpha - value added to the diagonal of the kernel matrix during fitting. Larger values correspond to increased 
# noise level in observations.
# gp_GHG = GaussianProcessRegressor(kernel = kernel_gp, alpha = 1e-3).fit(x_fit, y_fit_GHG)

#y_predict_GHG_GP, sigma_GHG_GP = gp_GHG.predict(x_predict, return_std=True)
y_predict_GHG_GP = gp_GHG.predict(x_predict)

# Kernel ridge regression
param_grid = {"alpha": [1e0, 1e-1, 1e-2, 1e-3], 
              "kernel": [ExpSineSquared(l,p) for l in np.logspace(0,3,10) for p in np.logspace(0,3,10)]}
kr_GHG = GridSearchCV(KernelRidge(), cv = 5, param_grid = param_grid)
kr_GHG.fit(x_fit, y_fit_GHG)

y_predict_GHG_kr = kr_GHG.predict(x_predict)

fig, (ax1) = plt.subplots(1, 1, figsize=(14, 6))
#ax1.plot(ncdata_tas_GHG_df, label = 'anthropogenic forcing')
#ax1.plot(xfit_GHG, yfit_GHG_OLS, 'r', label = 'OLS: slope = ' + '%.4f' % np.asscalar(ols_GHG.coef_))
#ax1.plot(xfit_GHG_predict, yfit_GHG_OLS_predict, 'g', label = 'OLS predict')
for i in range(ncdata_df.shape[1]): 
    ax1.plot(ncdata_df.iloc[:,i], label='_nolegend_')
ax1.plot(ncdata_df.mean(axis=1), 'k', lw = 4)
ax1.plot(x_predict, y_predict_GHG_GP, 'b', lw = 4, label = 'Gaussian processes predict')
ax1.fill(np.concatenate([x_predict, x_predict[::-1]]),
         np.concatenate([y_predict_GHG_GP.reshape(-1) - 1.9600 * sigma_GHG_GP,
                        (y_predict_GHG_GP.reshape(-1) + 1.9600 * sigma_GHG_GP)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
ax1.plot(x_predict, y_predict_GHG_kr, 'm', lw = 4, label = 'Kernel ridge regression')
ax1.set(xlabel = 'Year')
ax1.set(ylabel = 'Temperature anomaly (K)')
ax1.set_title('Global mean temperature anomaly')
ax1.legend()
plt.show()
```
