```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import netCDF4
from sklearn import linear_model
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, RationalQuadratic, ExpSineSquared, Matern, DotProduct, WhiteKernel, ConstantKernel

%matplotlib inline
```

```python
# RCP 4.5
dirFiles = "./../data/_DATA/CMIP5/1D/global/annual/tas/"
selFileContains = "rcp45"
ncdata_rcp45_df = pd.DataFrame()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

for f in os.listdir(dirFiles):
    if selFileContains in f:
        if not f.startswith("._"):
            temp_nc = netCDF4.Dataset(dirFiles + f)
            temp_ncdata = np.array(temp_nc.variables['tas']) - np.mean(temp_nc.variables['tas'])
            temp_ncdata = temp_ncdata.reshape(temp_ncdata.shape[0])
        
            temp_ncdata_df = pd.DataFrame(temp_ncdata, columns =[f])
            ncdata_rcp45_df = pd.concat( [ncdata_rcp45_df, temp_ncdata_df], axis = 1 )
        
times = temp_nc.variables['time']
dates = pd.to_datetime(netCDF4.num2date(times[:], times.units)).year
ncdata_rcp45_df.set_index(dates, inplace = True)

for i in range(ncdata_rcp45_df.shape[1]): 
    ax1.plot(ncdata_rcp45_df.iloc[:,i])
ax1.plot(ncdata_rcp45_df.mean(axis=1), 'k', lw = 4)
ax1.set(ylabel = "Temperature anomaly")
ax1.set_title("RCP 4.5")
    
# RCP 8.5
dirFiles = "./../data/_DATA/CMIP5/1D/global/annual/tas/"
selFileContains = "rcp85"
ncdata_rcp85_df = pd.DataFrame()

for f in os.listdir(dirFiles):
    if selFileContains in f:
        temp_nc = netCDF4.Dataset(dirFiles + f)
        temp_ncdata = np.array(temp_nc.variables['tas']) - np.mean(temp_nc.variables['tas'])
        temp_ncdata = temp_ncdata.reshape(temp_ncdata.shape[0])
        
        temp_ncdata_df = pd.DataFrame(temp_ncdata, columns =[f])
        ncdata_rcp85_df = pd.concat( [ncdata_rcp85_df, temp_ncdata_df], axis = 1 )

times = temp_nc.variables['time']
dates = pd.to_datetime(netCDF4.num2date(times[:], times.units)).year
ncdata_rcp85_df.set_index(dates, inplace = True)
ncdata_rcp85_df = ncdata_rcp85_df.drop(ncdata_rcp85_df.columns[[8,9]], axis = 1) # 8 and 9 have bad outliers

for i in range(ncdata_rcp85_df.shape[1]): 
    ax2.plot(ncdata_rcp85_df.iloc[:,i])
ax2.plot(ncdata_rcp85_df.mean(axis=1), 'k', lw = 4)
ax2.set(ylabel = "Temperature anomaly")
ax2.set_title("RCP 8.5")
```

### Gaussian processes for RCP 4.5

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4))

# predict starting from 2005
start = 1870
end_fit = 2005
end_predict = 2200

ncdata_train_df = ncdata_rcp45_df

x_fit = np.array(dates[ : end_fit - start + 1]).reshape(-1,1)
y_fit_RCP = ncdata_train_df.loc[start : end_fit].mean(axis = 1)
x_predict = np.linspace(start, end_predict, end_predict-start).reshape(-1,1)

kernel_GP = RBF(2)

gp_RCP = GaussianProcessRegressor(kernel = kernel_GP, alpha = 1e-2).fit(x_fit, y_fit_RCP)
y_predict_RCP_GP, sigma_RCP_GP = gp_RCP.predict(x_predict, return_std=True)

for i in range(ncdata_train_df.shape[1]): 
    ax1.plot(ncdata_train_df.iloc[:,i])
ax1.plot(ncdata_train_df.mean(axis=1), 'k', lw = 4)

ax1.plot(x_predict, y_predict_RCP_GP, 'b', lw = 4, label = 'Gaussian processes predict')
ax1.fill(np.concatenate([x_predict, x_predict[::-1]]),
         np.concatenate([y_predict_RCP_GP.reshape(-1) - 1.9600 * sigma_RCP_GP,
                        (y_predict_RCP_GP.reshape(-1) + 1.9600 * sigma_RCP_GP)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
ax1.axvline(x = end_fit,color = 'r')
ax1.set(xlabel = 'Year')
ax1.set(ylabel = 'Temperature anomaly (K)')
ax1.set_title('Global mean temperature anomaly (RCP 4.5)')

# predict starting from 2050
start = 1870
end_fit = 2050
end_predict = 2200

x_fit = np.array(dates[ : end_fit - start + 1]).reshape(-1,1)
y_fit_RCP = ncdata_train_df.loc[start : end_fit].mean(axis = 1)
x_predict = np.linspace(start, end_predict, end_predict-start).reshape(-1,1)

kernel_GP = RBF(2) + Matern(length_scale=2,nu=1) + WhiteKernel(1)

gp_RCP = GaussianProcessRegressor(kernel = kernel_GP, alpha = 1e-2).fit(x_fit, y_fit_RCP)
y_predict_RCP_GP, sigma_RCP_GP = gp_RCP.predict(x_predict, return_std=True)

for i in range(ncdata_train_df.shape[1]): 
    ax2.plot(ncdata_train_df.iloc[:,i])
ax2.plot(ncdata_train_df.mean(axis=1), 'k', lw = 4)

ax2.plot(x_predict, y_predict_RCP_GP, 'b', lw = 4, label = 'Gaussian processes predict')
ax2.fill(np.concatenate([x_predict, x_predict[::-1]]),
         np.concatenate([y_predict_RCP_GP.reshape(-1) - 1.9600 * sigma_RCP_GP,
                        (y_predict_RCP_GP.reshape(-1) + 1.9600 * sigma_RCP_GP)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
ax2.axvline(x = end_fit,color = 'r')
ax2.set(xlabel = 'Year')
ax2.set(ylabel = 'Temperature anomaly (K)')
ax2.set_title('Global mean temperature anomaly (RCP 4.5)')
```

### Gaussian processes for RCP 8.5

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4))

# predict starting from 2005
start = 1870
end_fit = 2005
end_predict = 2100

ncdata_train_df = ncdata_rcp85_df

x_fit = np.array(dates[ : end_fit - start + 1]).reshape(-1,1)
y_fit_RCP = ncdata_train_df.loc[start : end_fit].mean(axis = 1)
x_predict = np.linspace(start, end_predict, end_predict-start).reshape(-1,1)

kernel_GP = RBF(70) * DotProduct(50) #+ Matern(length_scale=2,nu=1)# + WhiteKernel(0.1)

gp_RCP = GaussianProcessRegressor(kernel = kernel_GP, alpha = 1e-2).fit(x_fit, y_fit_RCP)
y_predict_RCP_GP, sigma_RCP_GP = gp_RCP.predict(x_predict, return_std=True)

for i in range(ncdata_train_df.shape[1]): 
    ax1.plot(ncdata_train_df.iloc[:,i])
ax1.plot(ncdata_train_df.mean(axis=1), 'k', lw = 4)

ax1.plot(x_predict, y_predict_RCP_GP, 'b', lw = 4, label = 'Gaussian processes predict')
ax1.fill(np.concatenate([x_predict, x_predict[::-1]]),
         np.concatenate([y_predict_RCP_GP.reshape(-1) - 1.9600 * sigma_RCP_GP,
                        (y_predict_RCP_GP.reshape(-1) + 1.9600 * sigma_RCP_GP)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
ax1.axvline(x = end_fit,color = 'r')
ax1.set(xlabel = 'Year')
ax1.set(ylabel = 'Temperature anomaly (K)')
ax1.set_title('Global mean temperature anomaly (RCP 8.5)')

# predict starting from 2050
start = 1870
end_fit = 2050
end_predict = 2100

x_fit = np.array(dates[ : end_fit - start + 1]).reshape(-1,1)
y_fit_RCP = ncdata_train_df.loc[start : end_fit].mean(axis = 1)
x_predict = np.linspace(start, end_predict, end_predict-start).reshape(-1,1)

kernel_GP = RBF(100) * DotProduct(30) + Matern(length_scale=2,nu=1)  + WhiteKernel(0.1)

gp_RCP = GaussianProcessRegressor(kernel = kernel_GP, alpha = 1e-2).fit(x_fit, y_fit_RCP)
y_predict_RCP_GP, sigma_RCP_GP = gp_RCP.predict(x_predict, return_std=True)

for i in range(ncdata_train_df.shape[1]): 
    ax2.plot(ncdata_train_df.iloc[:,i])
ax2.plot(ncdata_train_df.mean(axis=1), 'k', lw = 4)

ax2.plot(x_predict, y_predict_RCP_GP, 'b', lw = 4, label = 'Gaussian processes predict')
ax2.fill(np.concatenate([x_predict, x_predict[::-1]]),
         np.concatenate([y_predict_RCP_GP.reshape(-1) - 1.9600 * sigma_RCP_GP,
                        (y_predict_RCP_GP.reshape(-1) + 1.9600 * sigma_RCP_GP)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
ax2.axvline(x = end_fit,color = 'r')
ax2.set(xlabel = 'Year')
ax2.set(ylabel = 'Temperature anomaly (K)')
ax2.set_title('Global mean temperature anomaly (RCP 8.5)')
```

```python
filename = './../data/_DATA/reg_glob/FORCING_CMIP5/global/ALLDATA_30May2010/RCP45_MIDYEAR_RADFORCING.xls'
df_rcp45_radforcing = pd.read_excel(filename, sheetname = 3, skiprows = 59, header = 0, index_col = 0)
df_rcp45_radforcing.index.name = 'YEAR'

filename = './../data/_DATA/reg_glob/FORCING_CMIP5/global/ALLDATA_30May2010/RCP85_MIDYEAR_RADFORCING.xls'
df_rcp85_radforcing = pd.read_excel(filename, sheetname = 3, skiprows = 59, header = 0, index_col = 0)
df_rcp85_radforcing.index.name = 'YEAR'
```

```python
fig, axes = plt.subplots(2, 2, figsize = (18, 10))

start = 1870
end = 2100

# RCP 4.5
df_rcp45_radforcing_subset = df_rcp45_radforcing.loc[start : end]

axes[0,0].plot(df_rcp45_radforcing_subset['TOTAL_ANTHRO_RF'], label = 'anthropogenic forcing')
axes[0,0].plot(df_rcp45_radforcing_subset['VOLCANIC_ANNUAL_RF'], label = 'volcanic forcing')
axes[0,0].plot(df_rcp45_radforcing_subset['SOLAR_RF'], label = 'solar forcing')
axes[0,0].plot(df_rcp45_radforcing_subset['VOLCANIC_ANNUAL_RF'] + df_rcp45_radforcing_subset['SOLAR_RF'], 
         label = 'natural forcing (volcanic + solar)')
axes[0,0].plot(df_rcp45_radforcing_subset['TOTAL_INCLVOLCANIC_RF'], 'k', label = 'all forcings')
axes[0,0].legend()
axes[0,0].set_title('Radiative forcing RCP 4.5')

ncdata_train45_df = ncdata_rcp45_df.loc[start : end]

x_fit = np.array(dates[start : end]).reshape(-1,1)
y_fit_RCP45 = ncdata_train45_df.mean(axis = 1)

ols_RCP45 = linear_model.LinearRegression()
ols_RCP45.fit(y_fit_RCP45.values.reshape(-1,1), df_rcp45_radforcing_subset['TOTAL_INCLVOLCANIC_RF'].values.reshape(-1,1))
radforcing_RCP45_predict_OLS = ols_RCP45.predict(y_fit_RCP45.values.reshape(-1,1))

ridge_RCP45 = linear_model.RidgeCV()
ridge_RCP45.fit(y_fit_RCP45.values.reshape(-1,1), df_rcp45_radforcing_subset['TOTAL_INCLVOLCANIC_RF'].values.reshape(-1,1))
radforcing_RCP45_predict_ridge = ridge_RCP45.predict(y_fit_RCP45.values.reshape(-1,1))

axes[0,1].plot(y_fit_RCP45, df_rcp45_radforcing_subset['TOTAL_INCLVOLCANIC_RF'], '.')
axes[0,1].plot(y_fit_RCP45, radforcing_RCP45_predict_OLS, 'g', lw = 4, label = 'OLS predict')
axes[0,1].plot(y_fit_RCP45, radforcing_RCP45_predict_ridge, 'r', lw = 4, label = 'Ridge predict')
axes[0,1].set(xlabel = 'Temperature anomaly')
axes[0,1].set(ylabel = 'Total radiative forcing')
axes[0,1].set_title('Radiative forcing vs. temp - RCP 4.5')
    
# RCP 8.5
df_rcp85_radforcing_subset = df_rcp85_radforcing.loc[start : end]

axes[1,0].plot(df_rcp85_radforcing_subset['TOTAL_ANTHRO_RF'], label = 'anthropogenic forcing')
axes[1,0].plot(df_rcp85_radforcing_subset['VOLCANIC_ANNUAL_RF'], label = 'volcanic forcing')
axes[1,0].plot(df_rcp85_radforcing_subset['SOLAR_RF'], label = 'solar forcing')
axes[1,0].plot(df_rcp85_radforcing_subset['VOLCANIC_ANNUAL_RF'] + df_rcp85_radforcing_subset['SOLAR_RF'], 
         label = 'natural forcing (volcanic + solar)')
axes[1,0].plot(df_rcp85_radforcing_subset['TOTAL_INCLVOLCANIC_RF'], 'k', label = 'all forcings')
axes[1,0].legend()
axes[1,0].set_title('Radiative forcing RCP 8.5')
    
ncdata_train85_df = ncdata_rcp85_df.loc[start : end]

x_fit = np.array(dates[start : end]).reshape(-1,1)
y_fit_RCP85 = ncdata_train85_df.mean(axis = 1)

ols_RCP85 = linear_model.LinearRegression()
ols_RCP85.fit(y_fit_RCP85.values.reshape(-1,1), df_rcp85_radforcing_subset['TOTAL_INCLVOLCANIC_RF'].values.reshape(-1,1))
radforcing_RCP85_predict_OLS = ols_RCP85.predict(y_fit_RCP85.values.reshape(-1,1))

ridge_RCP85 = linear_model.RidgeCV()
ridge_RCP85.fit(y_fit_RCP85.values.reshape(-1,1), df_rcp85_radforcing_subset['TOTAL_INCLVOLCANIC_RF'].values.reshape(-1,1))
radforcing_RCP85_predict_ridge = ridge_RCP85.predict(y_fit_RCP85.values.reshape(-1,1))

axes[1,1].plot(y_fit_RCP85, df_rcp85_radforcing_subset['TOTAL_INCLVOLCANIC_RF'], '.')
axes[1,1].plot(y_fit_RCP85, radforcing_RCP85_predict_OLS, 'g', lw = 4, label = 'OLS predict')
axes[1,1].plot(y_fit_RCP85, radforcing_RCP85_predict_ridge, 'r', lw = 4, label = 'Ridge predict')
axes[1,1].legend()
axes[1,1].set(xlabel = 'RCP temperature anomaly')
axes[1,1].set(ylabel = 'Total radiative forcing')
axes[1,1].set_title('Radiative forcing vs. temp - RCP 8.5')
```
