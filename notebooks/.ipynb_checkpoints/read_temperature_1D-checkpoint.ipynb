```python
import matplotlib.pyplot as plt
import netCDF4
import numpy as np
import pandas as pd
import sys
import os

from mpl_toolkits.basemap import Basemap

sys.path.append("./../src/")
from functions import * 

%matplotlib inline
```

#### 1. GFDL-CM3 - historicalNat and historicalGHG runs

```python
dirFiles = "./../data/mount/_DATA/CMIP5/1D/global/annual/tas/"
# historicalNat
nc_tas_Nat1 = netCDF4.Dataset(dirFiles + 'tas_mon_GFDL-CM3_historicalNat_r1i1p1_g025.nc')
nc_tas_Nat3 = netCDF4.Dataset(dirFiles + 'tas_mon_GFDL-CM3_historicalNat_r3i1p1_g025.nc')
nc_tas_Nat5 = netCDF4.Dataset(dirFiles + 'tas_mon_GFDL-CM3_historicalNat_r5i1p1_g025.nc')

ncdata_tas_Nat1 = np.array(nc_tas_Nat1.variables['tas'])
ncdata_tas_Nat1 = ncdata_tas_Nat1.reshape(ncdata_tas_Nat1.shape[0])

ncdata_tas_Nat3 = np.array(nc_tas_Nat3.variables['tas'])
ncdata_tas_Nat3 = ncdata_tas_Nat3.reshape(ncdata_tas_Nat3.shape[0])

ncdata_tas_Nat5 = np.array(nc_tas_Nat5.variables['tas'])
ncdata_tas_Nat5 = ncdata_tas_Nat5.reshape(ncdata_tas_Nat5.shape[0])

times = nc_tas_Nat1.variables['time']
dates = pd.to_datetime(netCDF4.num2date(times[:], times.units)).year

ncdata_tas_Nat_df1 = pd.DataFrame(ncdata_tas_Nat1, index = dates, columns = ['Temperature'])
ncdata_tas_Nat_df3 = pd.DataFrame(ncdata_tas_Nat3, index = dates, columns = ['Temperature'])
ncdata_tas_Nat_df5 = pd.DataFrame(ncdata_tas_Nat5, index = dates, columns = ['Temperature'])

alldata_historicalNat = [np.array(ncdata_tas_Nat_df1), np.array(ncdata_tas_Nat_df3), np.array(ncdata_tas_Nat_df5)]
mean_historicalNat = pd.DataFrame(np.mean(alldata_historicalNat, axis = 0), index = dates)

# historicalGHG
nc_tas_GHG1 = netCDF4.Dataset(dirFiles + 'tas_mon_GFDL-CM3_historicalGHG_r1i1p1_g025.nc')
nc_tas_GHG3 = netCDF4.Dataset(dirFiles + 'tas_mon_GFDL-CM3_historicalGHG_r3i1p1_g025.nc')
nc_tas_GHG5 = netCDF4.Dataset(dirFiles + 'tas_mon_GFDL-CM3_historicalGHG_r5i1p1_g025.nc')

ncdata_tas_GHG1 = np.array(nc_tas_GHG1.variables['tas'])
ncdata_tas_GHG1 = ncdata_tas_GHG1.reshape(ncdata_tas_GHG1.shape[0])

ncdata_tas_GHG3 = np.array(nc_tas_GHG3.variables['tas'])
ncdata_tas_GHG3 = ncdata_tas_GHG3.reshape(ncdata_tas_GHG3.shape[0])

ncdata_tas_GHG5 = np.array(nc_tas_GHG5.variables['tas'])
ncdata_tas_GHG5 = ncdata_tas_GHG5.reshape(ncdata_tas_GHG5.shape[0])

times = nc_tas_GHG1.variables['time']
dates = pd.to_datetime(netCDF4.num2date(times[:], times.units)).year

ncdata_tas_GHG_df1 = pd.DataFrame(ncdata_tas_GHG1, index = dates, columns = ['Temperature'])
ncdata_tas_GHG_df3 = pd.DataFrame(ncdata_tas_GHG3, index = dates, columns = ['Temperature'])
ncdata_tas_GHG_df5 = pd.DataFrame(ncdata_tas_GHG5, index = dates, columns = ['Temperature'])

alldata_historicalGHG = [np.array(ncdata_tas_GHG_df1), np.array(ncdata_tas_GHG_df3), np.array(ncdata_tas_GHG_df5)]
mean_historicalGHG = pd.DataFrame(np.mean(alldata_historicalGHG, axis = 0), index = dates)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

ax1.plot(ncdata_tas_Nat_df1, label = 'historical Nat 1')
ax1.plot(ncdata_tas_Nat_df3, label = 'historical Nat 3')
ax1.plot(ncdata_tas_Nat_df5, label = 'historical Nat 5')
ax1.plot(mean_historicalNat, 'k', lw = 2, label = 'mean historical Nat')
ax1.set(ylabel='Temperature (K)')
ax1.set_title('GFDL - historicalNat (1870-2005)')
ax1.legend()

ax2.plot(ncdata_tas_GHG_df1, label = 'historical GHG 1')
ax2.plot(ncdata_tas_GHG_df3, label = 'historical GHG 3')
ax2.plot(ncdata_tas_GHG_df5, label = 'historical GHG 5')
ax2.plot(mean_historicalGHG, 'k', lw = 2, label = 'mean historical GHG')
ax2.set(ylabel='Temperature (K)')
ax2.set_title('GFDL - historicalGHG (1870-2005)')
ax2.legend();
```

#### 2. GFDL-CM3 - RCP scenarios

```python
dirFiles = "./../data/mount/_DATA/CMIP5/1D/global/annual/tas/"

nc_tas_rcp45_1 = netCDF4.Dataset(dirFiles + 'tas_mon_GFDL-CM3_rcp45_r1i1p1_g025.nc')
nc_tas_rcp45_3 = netCDF4.Dataset(dirFiles + 'tas_mon_GFDL-CM3_rcp45_r3i1p1_g025.nc')
nc_tas_rcp45_5 = netCDF4.Dataset(dirFiles + 'tas_mon_GFDL-CM3_rcp45_r5i1p1_g025.nc')
nc_tas_rcp85_1 = netCDF4.Dataset(dirFiles + 'tas_mon_GFDL-CM3_rcp85_r1i1p1_g025.nc')

ncdata_tas_rcp45_1 = np.array(nc_tas_rcp45_1.variables['tas'])
ncdata_tas_rcp45_1 = ncdata_tas_rcp45_1.reshape(ncdata_tas_rcp45_1.shape[0])

ncdata_tas_rcp45_3 = np.array(nc_tas_rcp45_3.variables['tas'])
ncdata_tas_rcp45_3 = ncdata_tas_rcp45_3.reshape(ncdata_tas_rcp45_3.shape[0])

ncdata_tas_rcp45_5 = np.array(nc_tas_rcp45_5.variables['tas'])
ncdata_tas_rcp45_5 = ncdata_tas_rcp45_5.reshape(ncdata_tas_rcp45_5.shape[0])

ncdata_tas_rcp85_1 = np.array(nc_tas_rcp85_1.variables['tas'])
ncdata_tas_rcp85_1 = ncdata_tas_rcp85_1.reshape(ncdata_tas_rcp85_1.shape[0])

times = nc_tas_rcp45_1.variables['time']
dates = pd.to_datetime(netCDF4.num2date(times[:], times.units)).year

ncdata_tas_rcp45_df1 = pd.DataFrame(ncdata_tas_rcp45_1, index = dates, columns = ['Temperature'])
ncdata_tas_rcp45_df3 = pd.DataFrame(ncdata_tas_rcp45_3, index = dates, columns = ['Temperature'])
ncdata_tas_rcp45_df5 = pd.DataFrame(ncdata_tas_rcp45_5, index = dates, columns = ['Temperature'])

times = nc_tas_rcp85_1.variables['time']
dates = pd.to_datetime(netCDF4.num2date(times[:], times.units)).year

ncdata_tas_rcp85_df1 = pd.DataFrame(ncdata_tas_rcp85_1, index = dates, columns = ['Temperature'])

alldata_rcp45 = [np.array(ncdata_tas_rcp45_df1), np.array(ncdata_tas_rcp45_df3), np.array(ncdata_tas_rcp45_df5)]
mean_rcp45 = pd.DataFrame(np.mean(alldata_rcp45, axis = 0), index = dates)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

ax1.plot(ncdata_tas_rcp45_df1, label = 'RCP 4.5 - r1')
ax1.plot(ncdata_tas_rcp45_df3, label = 'RCP 4.5 - r3')
ax1.plot(ncdata_tas_rcp45_df5, label = 'RCP 4.5 - r5')
ax1.plot(mean_rcp45, 'b', lw = 3, label = 'mean RCP 4.5')
ax1.plot(ncdata_tas_rcp85_df1, label = 'RCP 8.5')
ax1.plot(mean_historicalNat, 'k', lw = 3, label = 'mean historical Nat')
ax1.plot(mean_historicalGHG, 'g', lw = 3, label = 'mean historical GHG')
ax1.set(ylabel='Temperature (K)')
ax1.set_title('GFDL - RCPs (1870-2100)')
ax1.legend()

ax2.plot(ncdata_tas_rcp45_df1.loc['1870':'2005'], label = 'RCP 4.5 - r1')
ax2.plot(ncdata_tas_rcp45_df3.loc['1870':'2005'], label = 'RCP 4.5 - r3')
ax2.plot(ncdata_tas_rcp45_df5.loc['1870':'2005'], label = 'RCP 4.5 - r5')
ax2.plot(mean_rcp45.loc['1870':'2005'], 'b', lw = 3, label = 'mean RCP 4.5')
ax2.plot(ncdata_tas_rcp85_df1.loc['1870':'2005'], label = 'RCP 8.5')
ax2.plot(mean_historicalNat, 'k', lw = 3, label = 'mean historical Nat')
#ax2.plot(mean_historicalGHG, 'g', lw = 3, label = 'mean historical GHG')
ax2.set(ylabel='Temperature (K)')
ax2.set_title('GFDL - RCPs (1870-2005)')
ax2.legend();
```

#### 3. MPI-ESM-LR

```python
dirFiles = "./../data/mount/_DATA/CMIP5/1D/global/annual/tas/"

# control runs
nc_tas_piControl = netCDF4.Dataset(dirFiles + 'tas_mon_MPI-ESM-LR_piControl_r1i1p1_g025.nc')
ncdata_tas_piControl = np.array(nc_tas_piControl.variables['tas'])
ncdata_tas_piControl = ncdata_tas_piControl.reshape(ncdata_tas_piControl.shape[0])

times = nc_tas_piControl.variables['time']
dates = pd.to_datetime(netCDF4.num2date(times[:], times.units)).year

ncdata_tas_piControl_df1 = pd.DataFrame(ncdata_tas_piControl, index = dates, columns = ['Temperature'])

alldata_piControl = [np.array(ncdata_tas_piControl_df1)]
mean_piControl = pd.DataFrame(np.mean(alldata_piControl, axis = 0), index = dates, columns = ['Temperature'])

# RCPs
nc_tas_rcp45_1 = netCDF4.Dataset(dirFiles + 'tas_mon_MPI-ESM-LR_rcp45_r1i1p1_g025.nc')
nc_tas_rcp45_2 = netCDF4.Dataset(dirFiles + 'tas_mon_MPI-ESM-LR_rcp45_r2i1p1_g025.nc')
nc_tas_rcp45_3 = netCDF4.Dataset(dirFiles + 'tas_mon_MPI-ESM-LR_rcp45_r3i1p1_g025.nc')

nc_tas_rcp85_1 = netCDF4.Dataset(dirFiles + 'tas_mon_MPI-ESM-LR_rcp85_r1i1p1_g025.nc')
nc_tas_rcp85_2 = netCDF4.Dataset(dirFiles + 'tas_mon_MPI-ESM-LR_rcp85_r2i1p1_g025.nc')
nc_tas_rcp85_3 = netCDF4.Dataset(dirFiles + 'tas_mon_MPI-ESM-LR_rcp85_r3i1p1_g025.nc')

ncdata_tas_rcp45_1 = np.array(nc_tas_rcp45_1.variables['tas'])
ncdata_tas_rcp45_1 = ncdata_tas_rcp45_1.reshape(ncdata_tas_rcp45_1.shape[0])

ncdata_tas_rcp45_2 = np.array(nc_tas_rcp45_2.variables['tas'])
ncdata_tas_rcp45_2 = ncdata_tas_rcp45_2.reshape(ncdata_tas_rcp45_2.shape[0])

ncdata_tas_rcp45_3 = np.array(nc_tas_rcp45_3.variables['tas'])
ncdata_tas_rcp45_3 = ncdata_tas_rcp45_3.reshape(ncdata_tas_rcp45_3.shape[0])

ncdata_tas_rcp85_1 = np.array(nc_tas_rcp85_1.variables['tas'])
ncdata_tas_rcp85_1 = ncdata_tas_rcp85_1.reshape(ncdata_tas_rcp85_1.shape[0])

ncdata_tas_rcp85_2 = np.array(nc_tas_rcp85_2.variables['tas'])
ncdata_tas_rcp85_2 = ncdata_tas_rcp85_2.reshape(ncdata_tas_rcp85_2.shape[0])

ncdata_tas_rcp85_3 = np.array(nc_tas_rcp85_3.variables['tas'])
ncdata_tas_rcp85_3 = ncdata_tas_rcp85_3.reshape(ncdata_tas_rcp85_3.shape[0])

times = nc_tas_rcp45_1.variables['time']
dates = pd.to_datetime(netCDF4.num2date(times[:], times.units)).year

ncdata_tas_rcp45_df1 = pd.DataFrame(ncdata_tas_rcp45_1, index = dates, columns = ['Temperature'])
ncdata_tas_rcp45_df2 = pd.DataFrame(ncdata_tas_rcp45_2, index = dates, columns = ['Temperature'])
ncdata_tas_rcp45_df3 = pd.DataFrame(ncdata_tas_rcp45_3, index = dates, columns = ['Temperature'])

times = nc_tas_rcp85_1.variables['time']
dates = pd.to_datetime(netCDF4.num2date(times[:], times.units)).year

ncdata_tas_rcp85_df1 = pd.DataFrame(ncdata_tas_rcp85_1, index = dates, columns = ['Temperature'])
ncdata_tas_rcp85_df2 = pd.DataFrame(ncdata_tas_rcp85_2, index = dates, columns = ['Temperature'])
ncdata_tas_rcp85_df3 = pd.DataFrame(ncdata_tas_rcp85_3, index = dates, columns = ['Temperature'])

alldata_rcp45 = [np.array(ncdata_tas_rcp45_df1), np.array(ncdata_tas_rcp45_df2), np.array(ncdata_tas_rcp45_df3)]
mean_rcp45 = pd.DataFrame(np.mean(alldata_rcp45, axis = 0), index = dates, columns = ['Temperature'])

alldata_rcp85 = [np.array(ncdata_tas_rcp85_df1), np.array(ncdata_tas_rcp85_df2), np.array(ncdata_tas_rcp85_df3)]
mean_rcp85 = pd.DataFrame(np.mean(alldata_rcp85, axis = 0), index = dates, columns = ['Temperature'])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))

ax1.plot(ncdata_tas_rcp45_df1, label = 'RCP 4.5 - r1')
ax1.plot(ncdata_tas_rcp45_df2, label = 'RCP 4.5 - r2')
ax1.plot(ncdata_tas_rcp45_df3, label = 'RCP 4.5 - r3')
ax1.plot(mean_rcp45, 'b', lw = 3, label = 'mean RCP 4.5')
ax1.plot(ncdata_tas_rcp85_df1, label = 'RCP 8.5 - r1')
ax1.plot(ncdata_tas_rcp85_df2, label = 'RCP 8.5 - r2')
ax1.plot(ncdata_tas_rcp85_df3, label = 'RCP 8.5 - r3')
ax1.plot(mean_rcp85, 'r', lw = 3, label = 'mean RCP 8.5')
ax1.plot(ncdata_tas_piControl_df1, 'k', label = 'piControl - r1')
ax1.set(ylabel = 'Temperature (K)')
ax1.set_title('MPI-ESM-LR (1870-2100)')
ax1.legend()

ax2.plot(ncdata_tas_rcp45_df1.loc['1870':'2005'], label = 'RCP 4.5 - r1')
ax2.plot(ncdata_tas_rcp45_df2.loc['1870':'2005'], label = 'RCP 4.5 - r2')
ax2.plot(ncdata_tas_rcp45_df3.loc['1870':'2005'], label = 'RCP 4.5 - r3')
ax2.plot(mean_rcp45.loc['1870':'2005'], 'b', lw = 3, label = 'mean RCP 4.5')
ax2.plot(ncdata_tas_rcp85_df1.loc['1870':'2005'], label = 'RCP 8.5 - r1')
ax2.plot(ncdata_tas_rcp85_df2.loc['1870':'2005'], label = 'RCP 8.5 - r2')
ax2.plot(ncdata_tas_rcp85_df3.loc['1870':'2005'], label = 'RCP 8.5 - r3')
ax2.plot(mean_rcp85.loc['1870':'2005'], 'r', lw = 3, label = 'mean RCP 8.5')
ax2.plot(ncdata_tas_piControl_df1.loc['1870':'2005'], 'k', label = 'piControl - r1')
ax2.set(ylabel = 'Temperature (K)')
ax2.set_title('MPI-ESM-LR (1870-2005)')
ax2.legend();
```
